KPI

Segun lo observado a lo alrgo de EDA, he podido notar ciertas tendencias, por lo que opte en trabajar en ver un aumento de un 2% en el acceso al servicio internet para el proximo trimestre cada 100 hogares, primero ordeno el dataframe 

```python
df_accesos_prov.info()
```

```python
df_crecimiento = df_accesos_prov.sort_values(by=['Provincia', 'Año', 'Trimestre'])
```

Funcion para calcular kpi

```python
def kpi(valor_actual, valor_anterior):
    
    if valor_anterior != 0:
        return round((((valor_actual - valor_anterior) / valor_anterior) * 100), 2)
    else:
        return 0 
```

```python
df_crecimiento['Accesos trimestre anterior'] = df_crecimiento.groupby('Provincia')['Accesos por cada 100 hogares'].shift(1)
```

lleno los valores nulos

```python
df_crecimiento['Accesos trimestre anterior'] = df_crecimiento['Accesos trimestre anterior'].fillna(0)
```

Aplico la funcion

```python
df_crecimiento['KPI'] = df_crecimiento.apply(lambda row: kpi(row['Accesos por cada 100 hogares'], row['Accesos trimestre anterior']), axis=1)
df_crecimiento['KPI'] = df_crecimiento['KPI'].fillna(0)
```

# KPI en wireless y en Fibra Optica

Se quiere alcanzar mayor tecnologia en otras zonas enfocandose en la tecnologia wireless y fibra Optica buscando la oportunidad de negocio, evalúo lo oobtenido.

Primero creo mi dataframe y lo ordeno

```python
df_wireless_fo = df_accesos_prov.drop(columns=['ADSL','Cablemodem','Otros'])
df_wireless_fo = df_wireless_fo.sort_values(by=['Provincia', 'Año', 'Trimestre'])
```

Creo mi columna Accesos de trimestre anterior

```python
df_wireless_fo['Anterior Fibra Optica'] = df_wireless_fo.groupby('Provincia')['Fibra óptica'].shift(1)
df_wireless_fo['Anterior Wireless'] = df_wireless_fo.groupby('Provincia')['Wireless'].shift(1)
```

Calculo de los kpi

```python
df_wireless_fo['KPI wireless'] = df_wireless_fo.apply(lambda row: kpi(row['Wireless'], row['Anterior Wireless']), axis=1)
```

```python
df_wireless_fo['KPI F0'] = df_wireless_fo.apply(lambda row: kpi(row['Fibra óptica'], row['Anterior Fibra Optica']), axis=1)
```

Reviso el comportamiento
